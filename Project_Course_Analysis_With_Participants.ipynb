```python
Institution_Participants = df.groupby(['Institution','Course Subject']).mean().reset_index()
snb.factorplot(x='Institution', y='Participants (Course Content Accessed)', hue='Course Subject', data = Institution_Participants, kind='bar')
```

```python
Year_Participants = df.groupby(['Year','Course Subject']).mean().reset_index()
snb.factorplot(x='Year', y='Participants (Course Content Accessed)', hue='Course Subject', data=Year_Participants, kind='bar')
```

```python
Videos_Played = df.groupby(['Year','Institution']).mean().reset_index()
snb.factorplot(x='Year', y='Participants (Course Content Accessed)', hue='Institution', data=Videos_Played, kind='bar')
```

```python
Certified_Percentage = df.groupby(['Year','Institution']).mean().reset_index()
snb.factorplot(x='Year', y='% Certified', hue='Institution', data=Certified_Percentage, kind='bar')
```
